# 3. Graph for Gender participated in Sales

```python
import matplotlib.pyplot as plt
carsd.groupby('Buyer Gender').count()
```

Now we can say that 4948 males and 5052 females were participated in this program.

```python
gen=['Male','Female']
frq=[4948,5052]
plt.bar(gen,frq,align='center',alpha=0.5, width=0.9,color=['red','y'],edgecolor='orange')
plt.title('Gender wise Buyers')
plt.xlabel('Gender')
plt.ylabel("No of pariticipents")
plt.show()
```

interpretation:<br/>
<span style="background-color:Gold">**It is very clear that Female buyers are greater than that of Male buyers**</span>
